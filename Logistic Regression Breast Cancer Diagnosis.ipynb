# Check for Highly Correlated features

- Checks for pearsons correlarion scores
- If two features are >0.90 correlated, one of them is dropped
- As high correlation = redundant information

x_reduced_df = pd.DataFrame(x_reduced, columns=x.columns[selector.get_support()])

corr_matrix = x_reduced_df.corr().abs()

# Dropping the lower triangle of the matri as they are symmetrical in a correltation matrix
upper = corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=1).astype(bool)==False)

# Find features with correlation greater than 0.90
to_drop = [column for column in upper.columns if any(upper[column]> 0.90)]

x_reduced_df = x_reduced_df.drop(columns=to_drop)

- Precision: TP/(TP+FP) - Correct predictions out of all predicted positives [P]
- Recall: TP/(TP+FN) - Correct positive predictions out of all actual positives [R]
- F1-Score: 2 x [P x R]/[P + R] - Harmonic mean of precision and recall
- Support: Count of each class, number of true instances for each class

# ROC-AUC

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
```

```python
# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random classifier line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
thresholds_list = [0.3, 0.5, 0.7, 0.9]
for threshold in thresholds_list:
    y_pred = (y_pred >= threshold).astype(int)
    print(f"\nThreshold: {threshold}")
    print(confusion_matrix(y_test, y_pred))
    print(f"Precision: {precision_score(y_test, y_pred):.2f}")
    print(f"Recall: {recall_score(y_test, y_pred):.2f}")
    print(f"F1-Score: {f1_score(y_test, y_pred):.2f}")
```

### What is sigmoid function?

- A sigmoid function is a mathematical function that exhibits a characteristic "S"-shaped curve, known as a sigmoid curve.
- It maps any real-valued number into a range between 0 and 1, making it useful for converting outputs into probabilities.
- In machine learning, the term "sigmoid function" often refers specifically to the logistic function, which is widely used as an activation function in artificial neural networks.
- This function helps the network learn non-linear relationships between the data and is particularly beneficial in binary classification problems.

- The logistic function, a common example of a sigmoid function, is defined by the formula y = 1/1+ e^-x

- It has a derivative that is bell-shaped and is invertible, with its inverse being the logit function.

- Sigmoid functions are also utilized in statistics as cumulative distribution functions, which go from 0 to 1.
- They are advantageous in various fields, including deep learning, binary classification, and modeling systems that saturate at large values of x.

- Logistic function: y = 1/1+ e^-x
